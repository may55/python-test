```python
from datetime import date,timedelta
from nsepy import get_history
```

```python
days_diff = 365
strt = date(2015,1,1)
ends = strt+timedelta(days=days_diff)
sbin = get_history(symbol='SBIN',
                    start= strt,
                    end=ends)
sbin = sbin.reset_index()
```
