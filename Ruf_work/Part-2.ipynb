```python
from datetime import date,timedelta
from nsepy import get_history
```

```python
days_diff = 365
strt = date(2015,1,1)
ends = strt+timedelta(days=days_diff)
sbin = get_history(symbol='SBIN',
                    start= strt,
                    end=ends)
sbin = sbin.reset_index()
```

```python
sbin.head()
```

```python
fig = figure(x_axis_type="datetime")
```

```python
fig.line(sbin.index, sbin['Close'], color='blue', alpha=0.5)
fig.circle(sbin.index, sbin.Close*sbin["price_shock_w/0_vol_shock"], size=4, legend='price shock without vol shock')
show(fig)
```
